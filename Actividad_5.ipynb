```python
#Seleccionamos sólo las columnas con datos cuantitativos
cuantitativas_estocolmo = estocolmo.select_dtypes(include=['number'])
cualitativas_estocolmo = estocolmo.select_dtypes(include=['object', 'category'])
cuantitativas_mexico = mexico.select_dtypes(include=['number'])
cualitativas_mexico = mexico.select_dtypes(include=['object', 'category'])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,8))
cuantitativas_estocolmo.plot(kind='box', vert=False)
plt.title('Valores Atípicos de Estocolmo')
plt.show()
```

```python
imfig = plt.figure(figsize=(15,8))
cuantitativas_mexico.plot(kind='box', vert=False)
plt.title('Valores Atípicos del dataframe')
plt.show()
```

```python
#Método aplicando cuartiles. Encuentro cuartiles 0.25 y 0.75
#ESCTOCOLMO
y = cuantitativas_estocolmo

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q1
iqr = percentile75 - percentile25

Limite_superior_iqr = percentile75 + 1.5*iqr
Limite_inferior_iqr = percentile25 - 1.5*iqr

print('Límite superior permitido\n',Limite_superior_iqr)
print('Límite inferior permidito\n',Limite_inferior_iqr)
```

```python
estocolmo_limpio = cuantitativas_estocolmo[(y<=Limite_superior_iqr)&(y>=Limite_inferior_iqr)]
estocolmo_limpio
#Obtenemos datos limpios del DataFrame
```

```python
#Corroboramos los valores del DataFrame LIMPIO
valores_nulos = estocolmo_limpio.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del DataFream con 'mean'
#Realizamos una copia del DataFrame
estocolmo_iqr = estocolmo_limpio.copy()
estocolmo_iqr = estocolmo_iqr.fillna(round(estocolmo_limpio.mean(),1))
estocolmo_iqr
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo
Estocolmo_limpio_ = pd.concat([cualitativas_estocolmo,estocolmo_iqr], axis= 1)
Estocolmo_limpio_
```

```python
valores_nulos = Estocolmo_limpio_.isnull().sum()
valores_nulos
```
